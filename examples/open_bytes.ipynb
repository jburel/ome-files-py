```python
%matplotlib inline
import os
import urllib
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import ome_files
```

```python
fn, _ = urllib.urlretrieve("http://downloads.openmicroscopy.org/images/OME-TIFF/2016-06/tubhiswt-3D.zip")
```

```python
with zipfile.ZipFile(fn) as zip:
    for name in zip.namelist():
        with open(os.path.basename(name), "w") as fo:
            fo.write(zip.read(name))
```

```python
reader = ome_files.OMETIFFReader()
reader.set_id("tubhiswt_C0.ome.tif")
pixels = reader.open_array(0)
plt.imshow(pixels, cmap="gray")
reader.close()
```
